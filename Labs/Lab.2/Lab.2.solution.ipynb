*Exercise 3:* Modify exercise 2, so that it takes a matrix of the form from exercise 1 and draws a tic-tac-tie board with "X"s and "O"s.  

````python
# Write you solution here
def gameboard(matrix,printing):
    size = len(matrix)
    if printing == True:
        print(" --- " * size)
    for row in matrix:
        row_len=len(row)
        if printing == True:
            text = "| "
        for i in range(row_len):
            element = row[i] 
            if element == 1:
                if printing == True:
                    text = text +  "X" + "  | "
            elif element == 2:
                if printing == True:
                    text = text +  "O" + "  | "
            elif element == 0:
                if printing == True:
                    text = text +  " " + "  | "
        if printing == True:            
            print(text)
            print(" --- " * size)
        
````

*Exercise 4:* Write a function that takes a `n` by `n` matrix representing a tic-tac-toe game, and returns -1, 0, 1, or 2 indicating the game is incomplete, the game is a draw, player 1 has won, or player 2 has one, respectively. Here are some example inputs you can use to test your code:

````python
# Write you solution here
import numpy as np

def across(matrix): # -1 incomplete,0 draw, 1 p1 wins, 2 p2 wins
    for rows in matrix: #check for winners
        checker = rows[0] # gets the first number to check if they are the same in list
        winner = True
        for num in rows:
            if num == 0: #if any empty spaces stop checking
                winner = False
                break 
            elif num != checker: #makes sure all the numbers in the row are the saem
                winner = False
                break  
        if winner and checker != 0: 
            return checker #returns 1 or 2 depending on who wins

    for rows in matrix: #incomplete game
        for num in rows:
            if num == 0:
                return -1  # Incomplete game

    return 0  # if nothing else than its a draw




def vertical(matrix):  # just rearanges tge matrix
    matrix2 = matrix
    matrix2=np.array(matrix2)
    matrix2 = matrix2.T  # Or matrix.transpose()

    list_final = []
    for i in matrix2:
        list_final.append(list(i))
    list_final=across(list_final)
    return  list_final  
    #print(list_final)


def diagonal(matrix):
    rows = len(matrix) #row count
    right=[]
    left=[]
    for i in range(rows):             #makes the left diagonal 
        left.append(matrix[i][i])
    
    for i in range(rows):
        right.append(matrix[i][rows - 1 - i])   #makes the right diagonal by using reverse index

    return [left, right] #return the diagonals

def game(matrix):

    left, right = diagonal(matrix)

    if across(matrix) == 1 or vertical(matrix) == 1 or across([left])==1 or across([right])==1: 
        print("Player 1 won")  # Winner
        return 1
    elif across(matrix) == 2 or vertical(matrix) == 2 or across([left])==2 or across([right])==2: 
        print("Player 2 won")  # Winner
        return 2
    elif across(matrix) == 0 or vertical(matrix) == 0 or across([left])==0 or across([right])==0:
        print("Draw")  # No winner
        return 0
    else:
        print("incomplete game")# No winner yet
        return -1
````

*Exercise 5:* Write a function that takes a game board, player number, and `(x,y)` coordinates and places "X" or "O" in the correct location of the game board. Make sure that you only allow filling previously empty locations. Return `True` or `False` to indicate successful placement of "X" or "O".

````python
# Write you solution
def play(matrix,player_num,xy,printing):  #0 is empty, 1 is X, 2 is O
    gameboard(matrix,printing)
    if isinstance(xy, str): #if xy is a str then just make it into a list, if not then make into string then list
        pos = list(xy)
    else:
        pos = list(str(xy))
    x=int(pos[0])
    y=int(pos[1])
    if player_num==1:      
        if matrix[y][x] == 0: #if spot is empty
            matrix[y][x] = 1  #place an X
            gameboard(matrix,printing)
            return True
        
        elif matrix[y][x] == 1 or 2: #place nothing, spot is occupied
            return False
        
    elif player_num==2:
        if matrix[y][x] == 0: #if spot is empty
            matrix[y][x] = 2   #place an O
            gameboard(matrix,printing)
            return True
        
        elif matrix[y][x] == 1 or 2: #place nothing, spot is occupied
            return False
        
````
