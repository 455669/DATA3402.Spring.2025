```python
#combine all the sahpes into 1 and then add the new function
import math
class shape:
    def area(self):
        raise NotImplementedError()

    def perimeter(self):
        raise NotImplementedError()

    def get_coordinates(self):
        raise NotImplementedError()

    def get_length(self):
        raise NotImplementedError()

    def get_width(self):
        raise NotImplementedError()

    def get_radius(self):
        raise NotImplementedError()
    
    def get_pairs(self, pairs=16): #added because its used by every shape
        raise NotImplementedError()
    
    def is_inside(self, x, y): #user inputs the x and y then we check
        raise NotImplementedError()
        
    def is_overlap(self, other_shape): #checks if shapes would overlap
        raise NotImplementedError()
        
class rectangle(shape):
    def __init__(self, length, width,x,y): #double underscore makes variables private
        self.__length=int(length)
        self.__width=int(width)
        self.__x=int(x)
        self.__y=int(y)
        
    def area(self):      #simple area calculation
        area=self.__length * self.__width
        return area
    
    def perimeter(self): #rectangel has 2 widths and lengths
        perim= (self.__length*2 ) + (self.__width *2)
        return perim
    
    def get_length(self): #simple return of values
        return self.__length
        
    def get_width(self):
        return self.__width
        
    def get_coordinates(self):
        return self.__x,self.__y  
    
    def get_pairs(self, pairs=16): #subdivide the perimeter and get points for each equidistant piece
        points = []
        perimeter = 2 * (self.__length + self.__width)
        subdivison = perimeter / pairs #tells us how many subdivisions there are in the rectangle
        distance = 0

        for i in range(pairs): #loop on the subdiviosns
            if distance < self.__length: #get the coordinates as we go along all sides of the rectangle
                x = self.__x + distance
                y = self.__y
            elif distance < self.__length + self.__width:
                x = self.__x + self.__length
                y = self.__y + (distance - self.__length)
            elif distance < 2 * self.__length + self.__width:
                x = self.__x + self.__length - (distance - self.__length - self.__width)
                y = self.__y + self.__width
            else:
                x = self.__x
                y = self.__y + self.__width - (distance - 2 * self.__length - self.__width)
            points.append((x, y))
            distance += subdivison #increment for each subdivion piece
        return points
    
    def is_inside(self, x, y): #self.__x and self.__y is the center so we need to calculate the edges of the shape
        half_length = self.__length / 2
        half_width = self.__width / 2

        left = self.__x - half_length
        right = self.__x + half_length
        top = self.__y - half_width
        bottom = self.__y + half_width
        
        #we return true if both coordinates are contained in our rectangle
        return (left <= x <= right) and (top <= y <= bottom) #making sure the coordinate is inbetween both edges
 
    def is_overlap(self, other_shape):
        if isinstance(other_shape, rectangle): #compares rectangle to rectabgle for overlap
            x1, y1 = self.get_coordinates()  # get all data
            x2, y2 = other_shape.get_coordinates()
            length1 = self.get_length()
            width1 = self.get_width()
            length2 = other_shape.get_length()
            width2 = other_shape.get_width()

            # Calculate the edges of the rectangles
            left1 = x1 - length1 / 2
            right1 = x1 + length1 / 2
            top1 = y1 - width1 / 2
            bottom1 = y1 + width1 / 2

            left2 = x2 - length2 / 2
            right2 = x2 + length2 / 2
            top2 = y2 - width2 / 2
            bottom2 = y2 + width2 / 2

            # Check to see if they overlap using the edges
            return not (right1 < left2 or
                        left1 > right2 or
                        bottom1 < top2 or
                        top1 > bottom2)

        elif isinstance(other_shape, circle): #checks the other shapes for overlap
            return other_shape.is_overlap(other_shape)
        elif isinstance(other_shape, triangle):
            return other_shape.is_overlap(other_shape)
        else:
            return False
    
    
class circle(shape):
    ##i could import  math and use math.pi
    def __init__(self, radius,x,y): #double underscore makes variables private
        self.__radius=float(radius)
        self.__x=int(x)
        self.__y=int(y) #i could also make a variable for pi but its not really needed
        
        
    def area(self):      #simple area calculation pi*r^2
        area = math.pi*(self.__radius**2) #using rounded for simplicity
        area=round(area, 2)
        return area
    
    def perimeter(self): #circle perimeter is 2*pi*r
        perim= 2*math.pi*self.__radius
        perim=round(perim, 2)
        return perim
    
    def get_radius(self): #simple return of values
        return self.__radius
        
    def get_coordinates(self):
        return self.__x,self.__y

    
    def get_pairs(self, pairs=16):
        points = []
        angle_increment = 2 * math.pi / pairs   # the subdivisons of a circle in 16 pieces
        for i in range(pairs):
            angle = i * angle_increment   #move along the subdivion pieces
            x = self.__x + self.__radius * math.cos(angle)   #get the coordinates of the subdivision
            y = self.__y + self.__radius * math.sin(angle)
            points.append((x, y))
        return points
    
    def is_inside(self, x, y):
        distance = math.sqrt((x - self.__x) ** 2 + (y - self.__y) ** 2) #distance formula for distance of point from center
        return distance <= self.__radius #the circle edges are equidistant so we just need to compare against radius

    def is_overlap(self, other_shape):
        if isinstance(other_shape, circle):  # Compare circle with circle
            x1, y1 = self.get_coordinates()
            x2, y2 = other_shape.get_coordinates()
            distance = math.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)
            return distance <= self.__radius + other_shape.get_radius()
        
        elif isinstance(other_shape, rectangle):  # If other_shape is rectangle
            return other_shape.is_overlap(other_shape)
        elif isinstance(other_shape, triangle):  # If other shape is triangle
            return other_shape.is_overlap(other_shape)
        else:
            return False
    
class triangle(shape):
    def __init__(self, side1, side2, side3, x, y):
        self.__side1 = float(side1)
        self.__side2 = float(side2)
        self.__side3 = float(side3)
        self.__x = int(x)
        self.__y = int(y)

    def area(self):
        semi = (self.__side1 + self.__side2 + self.__side3) / 2  # semi perimeter from herons formula
        area = math.sqrt(semi * (semi - self.__side1) * (semi - self.__side2) * (semi - self.__side3))
        area=round(area, 2)
        return area

    def perimeter(self): 
        perim = self.__side1 + self.__side2 + self.__side3 #rounded to keep simple
        perim=round(perim, 2)
        return perim

    def get_coordinates(self): #simple return of all the traingle values
        return self.__x, self.__y

    def get_side1(self):
        return self.__side1

    def get_side2(self):
        return self.__side2

    def get_side3(self):
        return self.__side3
    
    
    def get_pairs(self, pairs=16):
        points = []
        perimeter = self.__side1 + self.__side2 + self.__side3
        piece = perimeter / pairs   # equal subdivisions of traingle
        distance = 0 # 

        for i in range(pairs):
            if distance < self.__side1: # move along the sides of the triangle collecting points
                x = self.__x + distance
                y = self.__y
            elif distance < self.__side1 + self.__side2:
                x = self.__x + self.__side1
                y = self.__y + (distance - self.__side1)
            else:
                x = self.__x + self.__side1 - (distance - self.__side1 - self.__side2)
                y = self.__y + self.__side2
            points.append((x, y))
            distance += piece #slowly move along each piece until we get 16 point pairs
        return points
    
    def is_inside(self, x, y): #very hard
        def sign(p1, p2, p3):
            return (p1[0] - p3[0]) * (p2[1] - p3[1]) - (p2[0] - p3[0]) * (p1[1] - p3[1])

        v1 = (self.__x, self.__y)
        v2 = (self.__x + self.__side1, self.__y)
        v3 = (self.__x + (self.__side1 / 2), self.__y + self.area() / (self.__side1 / 2))
        pt = (x, y)

        d1 = sign(pt, v1, v2)
        d2 = sign(pt, v2, v3)
        d3 = sign(pt, v3, v1)

        has_neg = (d1 < 0) or (d2 < 0) or (d3 < 0)
        has_pos = (d1 > 0) or (d2 > 0) or (d3 > 0)

        return not (has_neg and has_pos)
    #trinagle orientation and size has to be considered aswell as the edge le
    def is_overlap(self, other_shape): #will not overlap
        if isinstance(other_shape, triangle):
            return True
        elif isinstance(other_shape, rectangle):
            return other_shape.is_overlap(other_shape)
        elif isinstance(other_shape, circle):
            return other_shape.is_overlap(other_shape)
        else:
            return False
```

```python
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        # Empty canvas is a matrix with element being the "space" character
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x,x+w):
            self.set_pixel(i,y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y,y+h):
            self.set_pixel(x,i, **kargs)
            
    def line(self, x1, y1, x2, y2, **kargs):
        slope = (y2-y1) / (x2-x1)
        for y in range(y1,y2):
            x= int(slope * y)
            self.set_pixel(x,y, **kargs)
            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))
```

```python
class foo:
    def __init__(self,a,b=None):
        self.a=a
        self.b=b
        
    def __repr__(self):
        return "foo("+repr(self.a)+","+repr(self.b)+")"
    
    def save(self,filename):
        f=open(filename,"w")
        f.write(self.__repr__())
        f.close()
        
   
def foo_loader(filename):
    f=open(filename,"r")
    tmp=eval(f.read())
    f.close()
    return tmp
```
